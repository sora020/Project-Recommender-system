```python
movie_df['production_companies'] = movie_df['production_companies'].fillna('[]').apply(ast.literal_eval)
movie_df['production_companies'] = movie_df['production_companies'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
b = movie_df.apply(lambda x: pd.Series(x['production_companies']),axis=1).stack().reset_index(level=1, drop=True)
b.name = 'companies'
```

```python
com_df = movie_df.drop('production_companies', axis=1).join(b)
```

```python
com_df
```

```python
com_df.info()
```

```python
com_sum = pd.DataFrame(com_df.groupby('companies')['revenue'].sum().sort_values(ascending=False))
com_sum.columns = ['Total']
com_mean = pd.DataFrame(com_df.groupby('companies')['revenue'].mean().sort_values(ascending=False))
com_mean.columns = ['Average']
com_count = pd.DataFrame(com_df.groupby('companies')['revenue'].count().sort_values(ascending=False))
com_count.columns = ['Number']

com_pivot = pd.concat((com_sum, com_mean, com_count), axis=1)
```

```python
com_pivot
```

```python
#Highest grossing movie companies
com_pivot.sort_values('Total',ascending=False)
```

```python
#Companies with most No. of movies
com_pivot.sort_values('Number',ascending=False)
```

```python
#Most successful movie Companies
com_pivot[com_pivot['Number']>=10].sort_values('Average',ascending=False)
```

```python
credits_df = pd.read_csv('credits.csv')
credits_df.head()
```

```python
movie_df = movie_df.drop([19730, 29503, 35587])
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan
```

```python
movie_df['id'] = movie_df['id'].apply(convert_int)
```

```python
movie_df['id'] = movie_df['id'].astype('int')
```

```python
movie_df.info()
```

```python
movie_df = movie_df.merge(credits_df, on='id')
movie_df.shape
```

```python
movie_df.info()
```

```python
movie_df['cast'] = movie_df['cast'].apply(ast.literal_eval)
movie_df['crew'] = movie_df['crew'].apply(ast.literal_eval)
```

```python
movie_df['cast_size'] = movie_df['cast'].apply(lambda x: len(x))
movie_df['crew_size'] = movie_df['crew'].apply(lambda x: len(x))
```

```python
movie_df['cast'] = movie_df['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
movie_df['director'] = movie_df['crew'].apply(get_director)
```

```python
s =movie_df.apply(lambda x: pd.Series(x['cast']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'actor'
cast_df = movie_df.drop('cast', axis=1).join(s)
```

```python
sns.set_style('whitegrid')
plt.title('Actors with the Highest Total Revenue')
cast_df.groupby('actor')['revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```

```python
plt.title('Directors with the Highest Total Revenue')
movie_df.groupby('director')['revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar', colormap='autumn')
plt.show()
```

```python
actor_list = cast_df.groupby('actor')['revenue'].count().sort_values(ascending=False)
actor_list = list(actor_list[actor_list >= 5].index)
director_list = movie_df.groupby('director')['revenue'].count().sort_values(ascending=False)
director_list = list(director_list[director_list >= 5].index)
```

```python
plt.title("Actors with Highest Average Revenue")
cast_df[cast_df['actor'].isin(actor_list)].groupby('actor')['revenue'].mean().sort_values(ascending=False).head(10).plot(kind='bar', colormap='Greens_r')
plt.show()
```

```python
plt.title("Directors with Highest Average Revenue")
movie_df[movie_df['director'].isin(director_list)].groupby('director')['revenue'].mean().sort_values(ascending=False).head(10).plot(kind='bar', colormap='autumn_r')
plt.show()
```

```python
success_df = cast_df[(cast_df['return'].notnull()) & (cast_df['revenue'] > 1e7) & (cast_df['actor'].isin(actor_list))]
pd.DataFrame(success_df.groupby('actor')['return'].mean().sort_values(ascending=False).head(10))
```

```python
success_df = movie_df[(movie_df['return'].notnull()) & (movie_df['revenue'] > 1e7) & (movie_df['director'].isin(director_list))]
pd.DataFrame(success_df.groupby('director')['return'].mean().sort_values(ascending=False).head(10))
```
